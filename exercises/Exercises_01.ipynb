**1**. Make a list of consisting of the square of the numbers between 1 and 100 (inclusive) that are not divisible by 3 or 5

- Use a for loop
- Use map and filter
- using `numpy` arrays and vectorized operations

````python
import numpy as np
````

````python
ls = []
for x in range(1,101):
    if x%3==0 or x%5==0:
        continue
    else:
        ls.append(x)
list(np.array(ls)**2)
````

````python
def check_divisors(x):
    if x%3 == 0 or x%5 == 0:
        return False
    else:
        return True

list(np.array(
    list((filter(check_divisors, 
                 range(1,101))))
)**2)
````

**3**. A Pythagorean triplet is a tuple of numbers $(a, b, c)$ such that $a^2 + b^2 = c^2$. Find all unique Pythagorean triplets where $a$ and $b$ are each less than 20.

````python
import math
triples = []
for a in range(1,20):
    aa = a*a
    for b in range(1,20):
        bb = b*b
        cc = aa + bb
        c = math.sqrt(cc)
        if c.is_integer():
            triples.append((a,b,int(c)))

triples
````

**5**. Read the contents of the file `foo.txt` and use it to populat a 3 $\times$ 3 `numpy` array. Write the `numpy` array back to a file `bar.txt` using tabs as column separators and newlines as row separators.

````python
foo_arr = np.array([])
with open('foo.txt','r') as f:
    for line in f:
        if foo_arr.shape[0]<1:
            foo_arr = np.array([int(i) for i in line if i.isdigit()])
        else:
            foo_arr = np.append(foo_arr,[int(i) for i in line if i.isdigit()])
            
foo_arr = foo_arr.reshape(3,3)
np.savetxt("bar.txt",foo_arr,delimiter="\t")
````
